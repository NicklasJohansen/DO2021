# I gang med kort over kommuner og uddannelsesinstitutioner

### Installer først geopandas og shapely

Først skal pakkerne [geopandas](https://geopandas.org/) og [shapely](https://shapely.readthedocs.io/en/stable/manual.html) installeres, der gør det muligt at arbejde med geodata med Pandas. Åbn *Anaconda Prompt* og skriv: 

`conda install geopandas`

Godkend installationen af øvrige pakker med `y`. Shapely installeres sammen med geopandas. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, Polygon, box
from shapely import affinity
```

### Importer datasæt til kommunekort

**NB: Brug ikke for meget tid på at forstå denne kode, men kig på datasættet *kommunekort*.**

Nedenstående kode genererer datasættet *kommunekort*, der bruges til at plotte et Danmarkskort med kommuneinddeling. Datasættet indeholder 2 variable: *kommune* og *geometry*. *kommune* er kommunenavnet og *geometry* beskriver kommunens afgrænsning og placering på Danmarkskortet. Datasættet indeholder én række for hver kommune. Geodata håndtereres med pakken *geopandas*. 

*Kilde: Kortdata fra [kortforsyningen.dk](http://www.kortforsyningen.dk)*

```python
kommunekort = gpd.read_file('Kommune.shp')

#Inkluder kun seneste kommunedata
kommunekort = kommunekort[kommunekort['fra']=='2007-01-01'].reset_index(drop=True) 

# Ret kommunenavne
kommunekort['kommune'] = kommunekort['navn'].str[:-8] # Fjern ' kommune' fra kommunenavnet
rettede_kommunenavne = {'Københavns':'København', 'Bornholms':'Bornholm', 'Nordfyns':'Nordfyn', 'Åbenrå':'Aabenraa', 'Århus':'Aarhus', 'Høje Tåstrup':'Høje-Taastrup', 'Lyngby-Tårbæk':'Lyngby-Taarbæk', 'Fåborg-Midtfyn':'Faaborg-Midtfyn'}
kommunekort['kommune'] = kommunekort['kommune'].replace(rettede_kommunenavne)

# Flyt Bornholm for at skabe et mere kompakt kort
bornholm = (kommunekort['kommune']=='Christiansø') | (kommunekort['kommune']=='Bornholm')
kommunekort.loc[bornholm,'geometry'] = kommunekort.loc[bornholm,'geometry'].apply(lambda x: affinity.translate(x, xoff=-180000, yoff=240000))

# Opret en ramme omkring Bornholm for at indikere flytning
graense_bornholm = box(*kommunekort.loc[bornholm,'geometry'].total_bounds)
graense_bornholm = affinity.scale(graense_bornholm, xfact=1.7, yfact=1.4, origin='center')
ramme_bornholm = pd.DataFrame(data={'kommune':['Frame (Bornholm)']})
ramme_bornholm = gpd.GeoDataFrame(ramme_bornholm, geometry=[graense_bornholm])

# Slet ubenyttede variable
kommunekort.drop(columns=['objectid', 'enhedid', 'enhedtype', 'art', 'SHAPE_Leng',
       'SHAPE_Area', 'fra', 'til', 'navn'], inplace=True)

kommunekort.head(3)
```

### Plot Danmarkskortet

Nedenstående kode plotter et kommuneinddelt Danmarkskort baseret på datasættet *kommunekort* (vha. pakken *matplotlib*). 

```python
ax = kommunekort.plot(edgecolor="none") # Definer plottet af kommuner
ramme_bornholm.plot(ax=ax, edgecolor="grey",facecolor="none") # Plot rammen omkring Bornholm
plt.axis('off') # Fjern akser
plt.show() # Print kortet
```

### Plot en variabel på kortet

Det er muligt at knytte en værdi til hver kommune i datasættet. Disse værdier kan plottes på kortet vha. en farveskala. 

Først skal en variabel tilføjes til datasættet *kommunekort* for hver kommune. Som eksempel tilføjes her variablen *tilfaeldig_var* med tilfældige værdier mellem 0 og 100.

```python
kommunekort_eksempel = kommunekort.copy()
kommunekort_eksempel['tilfaeldig_var'] = np.random.randint(0, 101, kommunekort_eksempel.shape[0])
kommunekort_eksempel.head(3)
```

Variablen kan nu plottes på Danmarkskortet med nedenstående kode. `column='tilfaeldig_var'` angiver variablen, der skal plottes. `cmap='plasma'` angiver et [matplotlib farvetema](https://matplotlib.org/stable/tutorials/colors/colormaps.html).  

```python
ax = kommunekort_eksempel.plot(edgecolor='none', column='tilfaeldig_var', cmap='plasma', legend=True)
ramme_bornholm.plot(ax=ax, edgecolor="grey",facecolor="none")
plt.axis('off')
plt.show()
```

### Placering af uddannelsesinstitutioner på kortet

**NB: Brug ikke for meget tid på at forstå denne kode, men kig på datasættet *institutioner_geo*.**

Nedenstående kode importerer et datasæt over uddannelsesinstitutioner og deres koordinater på kortet. 

```python
institutioner_sti = 'institutioner.xlsx'
institutioner = pd.read_excel(institutioner_sti)

# Opret korrekte koordinater for instutionerne
institutioner_geo = gpd.GeoDataFrame(institutioner, geometry=gpd.points_from_xy(institutioner['GEO_LAENGDE_GRAD'], institutioner['GEO_BREDDE_GRAD']))
institutioner_geo.crs = {'init' :'epsg:4326'} # Specifier Coordinate Reference System (CRS) for institutionernes koordinater
institutioner_geo = institutioner_geo.to_crs({'init': 'epsg:25832'}) # Konverter CRS for at matche kortet 

# Flyt institutioner på Bornholm til den nye placering af Bornholm på kortet
bornholm_inst = institutioner_geo['BEL_KOMMUNE_NAVN']=='Bornholms Regionskommune'
institutioner_geo.loc[bornholm_inst,'geometry'] = institutioner_geo.loc[bornholm_inst,'geometry'].apply(lambda x: affinity.translate(x, xoff=-180000, yoff=240000))

institutioner_geo.head(3)
```

Institutionerne kan nu plottes på kortet ved at inkludere linjen `institutioner_geo.plot(ax=ax, edgecolor="red")`

```python
ax = kommunekort.plot(edgecolor="none")
ramme_bornholm.plot(ax=ax, edgecolor="grey", facecolor="none")
institutioner_geo.plot(ax=ax, edgecolor="red")
plt.axis('off')
plt.show()
```
