```python
stringg = ""
for i in joe[joe.columns[0]]:
    try:
        stringg = stringg + i
    except:
        pass
stringg
```

```python
spg3_korpus=nltk.word_tokenize(stringg)
spg3_korpus
```

```python
from pprint import pprint
finder = nltk.collocations.BigramCollocationFinder.from_words(spg3_korpus)
pprint(finder.ngram_fd.most_common(5))
```

```python
stopwords = nltk.corpus.stopwords.words("danish")
stopwords
```

```python
finder = nltk.collocations.BigramCollocationFinder.from_words(spg3_korpus)
pprint(finder.ngram_fd.most_common(5))
```

```python
print(len(spg3_korpus))
words = [w for w in spg3_korpus if w.lower() not in stopwords]
print(len(words))
#fjerner 
```

```python
sletkomma = [",", ".","(",")"]
```

```python
print(len(words))
words = [w for w in words if w.lower() not in sletkomma]
print(len(words))
```

```python
finder = nltk.collocations.BigramCollocationFinder.from_words(words)
pprint(finder.ngram_fd.most_common(10))
```

```python
finder = nltk.collocations.TrigramCollocationFinder.from_words(words)
pprint(finder.ngram_fd.most_common(10))
```

```python
finder = nltk.collocations.QuadgramCollocationFinder.from_words(words)
pprint(finder.ngram_fd.most_common(10))
```

```python
#i disse collocations tegnes et billede af et DI, der gør det godt, men også kunne gøre det bedre. 
#Dansk byggeri fylder tilsyneladende en del. 
#Men det vurderes ikke, at man kan ikke sige det store ud fra dette.
```
