```python
def Region_til_tal(var):
    if var == 'Region Syddanmark':
        return 1
    if var == 'Region Nordjylland':
        return 2
    if var == 'Region Sjælland':
        return 3
    if var == 'Region Midtjylland':
        return 4
    if var == 'Region Hovedstaden':
        return 5
        
df['Region_tal'] = df['Virk region'].apply(Region_til_tal)
df
```

```python
def Brug_til_tal(var):
    if var == 'Intet brug':
        return 1
    if var == 'Lidt brug':
        return 2
    if var == 'Meget brug':
        return 3
        
df['Brug_tal'] = df['Brug'].apply(Brug_til_tal)
df
```

```python
sns.set_theme(style="darkgrid")
p_3 = sns.barplot(y="Brug_tal", x="Har besvaret NPM2020", data=df, ci=None, saturation=1)
```

```python
#definerer variabel NPM 2020 til at være tal.

def Dikotom_NPM(var):
    if var == 'Nej':
        return 0
    else:
        return 1
        
df['NPM2020_tal'] = df['Har besvaret NPM2020'].apply(Dikotom_NPM)
df
```
